```python
import pandas as pd
import os
import plotly.express as px
```

```python
df = pd.read_excel('Final extract.xlsx', sheet_name='DATA BRUT')
print(df)
```

```python
df['Year']
start_date = 1970
end_date = 2021
```

```python
#df['Year']=pd.to_datetime((df['Year']))
df=df.loc[(df['Year']>start_date) & (df['Year']<=end_date)]

# Prepare country group data / count the number of occurrences for each year
counts = df.groupby(['Year','Country']).count().reset_index()
print(counts)
```

```python
#CHOROPLETH settings in animation / Date mode. These lines allow us to create an animation that automatically scrolls through the years, as well as the association of the different colors on our scale (from 0 to 43).
fig = px.choropleth(counts, locations='Country',locationmode = "country names",hover_name='Country', animation_frame='Year', color='Disaster Type', 
                    color_continuous_scale=px.colors.sequential.Plasma, range_color=(0,43))

#Title and presentation
fig.update_layout(title_text=('Nb of extreme events'), title_x= 1, geo=dict(showframe=False, showcoastlines = False,))

#ligne de code demandant l'ouverture du fichier automatiquement lorsque cette ligne de code est run
fig.write_html('Extrem events map.html', auto_open=True)
fig
```
